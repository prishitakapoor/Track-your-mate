```python
from keras.applications.resnet50 import ResNet50
from keras.models import Model
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
import keras
from keras.layers import Dense, Dropout, Concatenate, Conv2D, Flatten, MaxPooling2D, Reshape, Input
import os as os
from PIL import Image
from tqdm import tqdm
import tensorflow as tf

import random

```

```python
basemodel = ResNet50(include_top = False, input_shape = (224, 224, 3))
x = basemodel.output
x = Conv2D(128, (3, 3), strides = 2)(x)
x = Conv2D(128, (3, 3), strides = 1)(x)
x = Reshape((128, ))(x)
basemodel = Model(inputs = basemodel.input, outputs = x)
basemodel.summary()


def create_model():
    
  anc = Input((224, 224, 3))
  pos = Input((224, 224, 3))
  neg = Input((224, 224, 3))
  
  anc_enc = basemodel(anc)
  pos_enc = basemodel(pos)
  neg_enc = basemodel(neg)
  
  out = Concatenate()([anc_enc, pos_enc, neg_enc])
  print(out.shape)
  
  model = Model(inputs = [anc, pos, neg], outputs = out)
  return model

```

```python
def score(x,y):
    return np.sum((x - y)**2)
```

```python
def getClass(x, all_enc, basemodel):
    encX = basemodel.predict(np.expand_dims(x, axis=0))
    min_score = np.inf
    categ = -1
    for idx, i in enumerate(all_enc.values()):
        a = score(encX,i)
        if min_score>a:
            min_score = a
            categ = idx
    print(a,list(all_enc.keys())[categ])
    return list(all_enc.keys())[categ]
```

```python
from firebase import firebase
import json
import time

import mxnet as mx
from mtcnn_detector import MtcnnDetector
import cv2
import os
import time
from sklearn.ensemble import RandomForestClassifier
import numpy as np

fire = firebase.FirebaseApplication('https://trackmate-c66f6.firebaseio.com/',None)

profEntryDict = {"Shivansh" : "03JRluOppbajITwzJYMaJD9gUEC3", "Ritik" : "BfbAAGmdaLPIVzqdafcdxolYcA72"}
#detail = {'email':'rkala@iiita.ac.in','location':location,'name':name,'nextAvailable':"Nothing to show",
#              'Time':str(time.strftime("%I:%M")),'online':False,'type':0}

def plot_prof (name, fire,location = "CC-3"):
    if(name == "No_one") : return
    result = fire.put('/professor/'+profEntryDict[name],'location',location)
    result = fire.put('/professor/'+profEntryDict[name],'time',str(time.strftime("%I:%M")) )
    print('updated' + str(name))
```

```python

detector = MtcnnDetector(model_folder='model', ctx=mx.cpu(0), num_worker = 4 , accurate_landmark = False)
model = RandomForestClassifier(n_estimators = 200, max_depth = 10)


dataset_path = './facedata/'
face_data = []
face_label = []

id = 0 #ID of exisiting data
names = {} #Dictionary for id- label mapping

#Data Preperation
for x in os.listdir(dataset_path):
	if(x.endswith('.npy')):
		x_name = x.replace(".npy","")
		names[id] = x_name
		data_item = np.load(dataset_path+x)
		face_data.append(data_item)

		id_label = id*np.ones((data_item.shape[0],))
		id += 1
		face_label.append(id_label)

X = np.concatenate(face_data,axis=0)
Y = np.concatenate(face_label, axis=0)
model.fit(X,Y)
#Read and Predict
cam = cv2.VideoCapture(0)
sample_face_data = []
```

```python
iter = 0
cache10 = {"Ritik" : 0, "Shivansh" : 0, "No_one" : 0}
while True:
    if (iter%5 == 0):
        cache10 = dict.fromkeys(cache10, 0)
        iter = 0
    ret,img = cam.read()
    iter += 1
    #Generally use Gray Fraqme to save memory
    results = detector.detect_face(img)

    if results is not None:
        total_boxes = results[0]
        points = results[1]
        # extract aligned face chips
        chips = detector.extract_image_chips(img, points, 144, 0.37)
        for i, chip in enumerate(chips):
            #prediction = getClass(chip,all_enc,basemodel)
            prediction = model.predict(chip.reshape(1,-1))
            if(prediction != None):
                name = names[int(prediction)]
                #name = prediction
                cache10[name] += 1
                #print(name + ' Score: ' + str(model.predict_proba(chip.reshape(1,-1))))
                if (cache10[name] > 2):
                    plot_prof(name, fire)
                    print("#####FOUND######" + name)

            
            
            
    cv2.imshow("Image",img)
    key_pressed = cv2.waitKey(1)
    if(key_pressed==ord('q')):
        break

cam.release()
```
